```python
import random
title = "WELCOME TO IPL 2022"
print(title.center(300))

ipl_team = ["CSK","MI","RCB","RR","LSG","GT","KXIP","KKR","SRH","DC"]
score_list = [0,1,2,4,6,"WICKET","NO BALL","WIDE","WICKET",6,0]

for team in ipl_team:
    print(team,end=" ")

my_team = input("\n\nEnter your team name : ").upper()
opp_team = ""

my_team_score = 0
opp_team_score = 0

m_ball = 0
o_ball = 0 

m_wicket = 0
o_wicket = 0

batting_team = ""

menu = """ 
                    press 1 for BAT
                    press 2 for BOWL
                """
def game():    
    global my_team,opp_team    
    if my_team in ipl_team:
        ipl_team.remove(my_team)
        opp_team = random.choice(ipl_team)
        print("\n MY TEAM : ",my_team)
        print("\n OPP. TEAM : ",opp_team)
    else:
        print("Invalid Name")


def toss_time(my_team,opp_team):
    global menu,batting_team
    print("\n\n TOSS (H/T) ")
    choice = input("Enter your choice (H/T) : ")
    toss = ("H","T")
    actual_toss = random.choice(toss)
    if choice==actual_toss:
        print(f"{my_team} WON THE TOSS")
        print(menu)
        
        selection = int(input("Enter your choice : "))
        
        if selection == 1:
            print(f"{my_team} choose bat first ")
            batting_team = my_team
            
        else:
            print(f"{my_team} choose bowl first ")
            batting_team = opp_team
            
    else:
        selection = random.randint(1,2)
        if selection == 1:
            batting_team = opp_team
            print(f"{opp_team} choose bat first ")
        else:
            batting_team = my_team
            print(f"{opp_team} choose bowl first ")
    print("---> batting team : ",batting_team)
        
def score_board(my_team,opp_team,batting_team):
    global my_team_score,opp_team_score,score_list,m_ball,o_ball,m_wicket,o_wicket
    if batting_team == my_team:
        while m_ball<10:
            m_ball += 1
            input()
            score = random.choice(score_list)
            if score=="WICKET":
                m_wicket += 1
                if m_wicket == 2:
                    print(f"{my_team} {my_team_score}/{m_wicket} ({m_ball} ball)  : {score}")
                    m_ball = 11
                else:
                    print(f"{my_team} {my_team_score}/{m_wicket} ({m_ball} ball)  : {score}")
            elif score=="WIDE" or score=="NO BALL":
                m_ball -= 1
                my_team_score += 1
                print(f"{my_team} {my_team_score}/{m_wicket} ({m_ball} ball)  : {score}")
            else:
                my_team_score += score
                print(f"{my_team} {my_team_score}/{m_wicket} ({m_ball} ball)  : {score}")
    else:
        while o_ball<10:
            o_ball += 1
            input()
            score = random.choice(score_list)
            if score=="WICKET":
                o_wicket += 1
                if o_wicket == 2:
                    print(f"{opp_team} {opp_team_score}/{o_wicket} ({o_ball} ball)  : {score}")    
                    o_ball=11
                else:
                    print(f"{opp_team} {opp_team_score}/{o_wicket} ({o_ball} ball)  : {score}")    
            elif score=="WIDE" or score=="NO BALL":
                o_ball -= 1
                opp_team_score += 1
                print(f"{opp_team} {opp_team_score}/{o_wicket} ({o_ball} ball)  : {score}")
            else:
                opp_team_score += score
                print(f"{opp_team} {opp_team_score}/{o_wicket} ({o_ball} ball)  : {score}")



game()
toss_time(my_team,opp_team)
play = 0
# print("---> play",play)
# print("---> bating name ",batting_team)
while play<=2:
    if batting_team==my_team:
        score_board(my_team,opp_team,my_team)
        batting_team = opp_team
        # print("--------> else -- bat",batting_team)
    else:
        score_board(my_team,opp_team,opp_team)
        batting_team = my_team
        # print("--> else -- bat",batting_team)
    play+=1 

if my_team_score>opp_team_score:
    print(f"{my_team} won the match !!")
else:
    print(f"{opp_team} won the match !!")
```
